**4.DEALING WITH NULL VALUES**





---



```python
data1['AMT_ANNUITY']=data1['AMT_ANNUITY'].ffill(axis = 0)
data1['AMT_ANNUITY'].isnull().sum()
```

```python
data1['AMT_ANNUITY'].describe()
```


**Removing Columns with an Threshold value**




```python
null_percent=((data1.isnull().sum())/(len(data1)))*100
print(null_percent)
```

```python
null_col=null_percent.index[null_percent>40]
len(null_col)
```

```python
data1=data1.drop(null_col,axis=1)
data1
```

**Removing Columns using drop() function**

```python
col=['AMT_REQ_CREDIT_BUREAU_HOUR','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR']
data1=data1.drop(col,axis=1)
data1
```

```python
col_rem=data1.filter(regex='^FLAG',axis=1).columns
col_rem
```

```python
data1=data1.drop(col_rem,axis=1)
data1
```

```python
col_rem1=['WEEKDAY_APPR_PROCESS_START','NAME_TYPE_SUITE','HOUR_APPR_PROCESS_START']
data1=data1.drop(col_rem1,axis=1)
data1
```

```python
data1.columns
```

```python
data1=data1.drop(['EXT_SOURCE_2','EXT_SOURCE_3'],axis=1)
data1
```

```python
data1.columns
```

```python
data1.isnull().sum()
```

```python
data1=data1.drop(['DAYS_LAST_PHONE_CHANGE'],axis=1)
data1
```

**6.IDENTIFYING DATA IMBALANCE**

---



```python
Imbalance = data1["TARGET"].value_counts().reset_index()
count_0 = Imbalance.iloc[0]["TARGET"]
count_1 = Imbalance.iloc[1]["TARGET"]
count_0_perc = round(count_0/(count_0+count_1)*100,2)
count_1_perc = round(count_1/(count_0+count_1)*100,2)
print('Ratios of imbalance in percentage: %.2f and %.2f'%(count_0_perc,count_1_perc))
print('Ratios of imbalance in relative %.2f : 1 (approx)'%(count_0/count_1))
```
